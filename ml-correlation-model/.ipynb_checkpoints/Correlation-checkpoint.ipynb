### Visualizing the Forex Data
I put all of the necessary import statements in this chunk.

```python
import os 
import random
from pandas import read_csv
from matplotlib import pyplot as plt
import numpy as np

span = 30 #span = how many to skip over, i.e get once every x times
path = ['ETFs', 'Forex Data']

forex_vals, forex_axis = [], []
forex = read_csv(path[1] + '/eurusd_hour.csv', header=0, index_col=0, parse_dates=True, 
                 usecols = ['Date', 'BC'], skiprows=lambda x: x % (span * 23)) # multiply by 23 to get one day
```

```python
fig1 = plt.figure(1)
fig2 = plt.figure(2)
ax3 = fig3.add_subplot(111)
ax4 = fig4.add_subplot(111)

ax1.scatter(forex.index, forex.pct_change())
forex_vals.extend([e[0] for e in forex.pct_change().values])
forex_axis = list(range(0, span * len(forex_vals), span))  # time axis should be every "span" times
forex_p = np.polyfit(forex_axis[1:], forex_vals[1:], 5)
forex_model = np.poly1d(forex_p)
forex_xp = np.linspace(0, span * len(forex_vals))
ax2.plot(forex_xp, forex_model(forex_xp))
```

The model definitely isn't perfect, and could probably use some more tuning. One thing that I think that could use modification is the scaling on the x-axis. The second plot for the model is shifted slightly to the left. I tried to approximate the x-axis change by having increments that correspond to the variable span, but this may be somewhat inaccurate. Further development could probably help fix this issue. 

### Processing the ETF Data
Since the .csv data for each ETF is in a separate folder, I grabbed all the file names and kept them in a list, which I could go through later. This was made possible by using the "listdir" function. 

```python
etfs = os.listdir(path[0])
```

After getting all the file names, it's time to process them. 

```python
fig3 = plt.figure(3)
fig4 = plt.figure(4)
ax3 = fig3.add_subplot(111)
ax4 = fig4.add_subplot(111)
fig5 = plt.figure(5)
ax5 = fig5.add_subplot(111)
added, x, y = {}, [], []
random.seed() 
num_files = 5 
for i in range (num_files):
    r = random.randrange(0, len(etfs) - 1)
    if r not in added.keys():
        etf = read_csv(path[0] + '/' + etfs[r], header=0, index_col=0, parse_dates=True, 
                       usecols = ['Date', 'Close'], skiprows=lambda x: x % span)
        ax1.scatter(etf.index, etf.pct_change().values)
        y.extend([e[0] for e in etf.pct_change().values])
        x = list(range(0, span * len(y), span)) # time axis should be every "span" times
        if len(x) == 0 or len(y) == 0:
            continue
        etf_p = np.polyfit(x[1:],y[1:], 5)
        etf_model = np.poly1d(etf_p)
        etf_xp = np.linspace(0, span * len(y))
        ax2.plot(etf_xp,etf_model(etf_xp))
        ax5.plot(forex_xp, (1 / len(y)) * np.power(etf_model(etf_xp) - forex_model(forex_xp), 2))
        y, x = [], []
    else: 
        r = random.randrange(0, len(etfs) - 1)
```

In this code snippet, what I do is read in a random .csv file for an ETF, and plot its percentage change over time. Each file is only read once. One idea for this snippet that I had was to randomly select a number of ETFs, plot them, and do this sequence a certain number of times, to see if most ETFs had a similar pattern. Right now, I have just been rerunning the cell to see different types of plots and spot trends. I think with more time, I might be able to make a more sophisticated way of doing this, but this is what I have for now. 

Also, I ended up trying to fit a polynomial to each ETF. I ended up deciding to use a degree of 5 for the polynomial, which I thought did the best job after many tests. If the degree is too high, there can definitely be overfitting, and if it is too low, there is a risk of not representing the data accurately. With more time and development, I expect to figure out a way to systematically decide which degree polynomial to use. 

Lastly, I performed a calculation where I calculated the Mean Squared Error. This should let me know the areas in which the model is not working well, and where further testing and tuning is needed.
